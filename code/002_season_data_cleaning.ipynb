```python
import pandas as pd

season_data = []
multi_team_keys = ["TOT", "2TM", "3TM", "4TM", "5TM"]
for year in range(1974, 2027):
    
    df_year = pd.read_csv(f'../data/Player Positions Data/Player Positions_{year}.csv')

    # Find players who have one of these aggregate team rows
    agg_players = df_year[df_year["Team"].isin(multi_team_keys)]["Player"].unique()

    # Keep:
    # 1. the aggregate row itself (TOT / 2TM / 3TM / etc.)
    # 2. players without any aggregate row
    df_year = df_year[
        (df_year["Team"].isin(multi_team_keys)) |
        (~df_year["Player"].isin(agg_players))
    ].copy()

    df_year = df_year.reset_index(drop=True)
    season_data.append(df_year)

df = pd.concat(season_data)
df.head()
```
